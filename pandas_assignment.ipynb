<h5> Q4: Create a dataframe containing 10 countries with the highest population. </h5>

```python
top_10_population = countries_df[['location', 'population']].sort_values(by = ['population'], ascending = False).head(10)
print(f'The top 10 population are: ')
top_10_population
```

<h5> Q5: Add a new column in countries_df to record the overall GDP per country (product of population & per capita GDP). </h5>


```python
countries_df['gdp'] = countries_df['population'] * countries_df['gdp_per_capita']
countries_df['gdp']
```

**Create a dataframe containing 10 countries with the lowest GDP per capita, among the counties with population greater than 100 million.**

```python
countries_population_more_than_100_million = countries_df[countries_df['population'] > 100000000]
countries_with_lowest_gdp = countries_population_more_than_100_million.sort_values(by = 'gdp_per_capita', ascending = True).head(10)
print('The top 10 countries with lowest gdp_per_capita and population more than 100 million: ')
countries_with_lowest_gdp
```
