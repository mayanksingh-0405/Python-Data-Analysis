#### First we import all the required libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Line Chart

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples)
```

<h5> Adding axes </h5>

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, yield_apples)
```

```python
plt.plot(years, yield_apples)
plt.xlabel('Years')
plt.ylabel('Yield of Apples')
```

<h5> Adding multiple lines </h5>

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Years')
plt.ylabel('Yield over the years')
```

<h5> Chart Title and Legend </h5>

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Years')
plt.ylabel('Yield over the years')
plt.title('Crops yield in Kanto')
plt.legend(['Apples', 'Oranges'])
```

<h5> Changing figure size </h5>

```python
plt.figure(figsize=(4,4))
plt.plot(years, oranges)
plt.title('Yield of Oranges')
```

```python
sns.set_style('darkgrid')
```

```python
plt.plot(years, apples, 's--b')
plt.plot(years, oranges, 'o--r')
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('Crop yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```

<h4> Scatter Plot </h4>

```python
# Loading data into pandas dataframe. 
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df
```

```python
flowers_df.shape
```

How many species are there in the table.

```python
flowers_df.species.unique()
```

```python
plt.scatter(flowers_df.sepal_length, flowers_df.sepal_width)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('Scatter Plot ')
```

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, s=100)
```

<h5> Plotting maps using Pandas dataframe </h5>

```python
plt.title('Sepal Dimension')
sns.scatterplot(x='sepal_length',
                y= 'sepal_width',
                hue = 'species',
                s = 100,
                data= flowers_df)
```

<h5> Multiple Histogram </h5>

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
```

```python
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
         bins = np.arange(2,5,0.5),
         stacked=True)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
```

<h5> Bar Plots with Averages </h5>

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df
```

<h5> Method 1 </h5>

```python
x = tips_df.groupby('day')['total_bill'].mean()
days = tips_df.day.unique()
colours = ['red', 'green', 'blue', 'purple']
```

```python
plt.bar(days,x, color = colours)
```

<h5> Method 2 </h5>

```python
colors = sns.color_palette('husl', n_colors=len(tips_df['day'].unique()))
sns.barplot(x='day', y = 'total_bill', data = tips_df, palette = colors)
```

```python
sns.barplot(x='day', y = 'total_bill', hue = 'sex', data = tips_df, palette=colors)
```

We can even switch the graph.

```python
sns.barplot(x='total_bill', y= 'day', hue= 'sex', data = tips_df, palette=colors)
```

<h4> Plotting multiple charts in a grid </h4>

```python
fig, axes = plt.subplots(2,3, figsize = (16,8))

# First Graph
axes[0,0].plot(years, apples, 's--b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Years')
axes[0,0].set_ylabel('Yields (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges'])
axes[0,0].set_title('Crop yields in Kanto')


# Second Graph
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=100, 
                ax=axes[0,1])

# Third Graph
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, virginica_df.sepal_width, versicolor_df.sepal_width],
               bins = np.arange(2,5,0.5),
               stacked = True)
axes[0,2].legend(['Setosa', 'Virginica', 'Versicolor'])


# Fourth Graph
axes[1,0].set_title('Restraunt Bills')
sns.barplot(x='day', y='total_bill', hue = 'sex', data = tips_df, ax = axes[1,0])

# Fifth Graph
axes[1,1].set_title('Fligth Traffix')
sns.heatmap(fligths_df, ax = axes[1,1])

# Sixth Graph
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])
```
