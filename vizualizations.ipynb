#### First we import all the required libraries.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Line Chart

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples)
```

<h5> Adding axes </h5>

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, yield_apples)
```

```python
plt.plot(years, yield_apples)
plt.xlabel('Years')
plt.ylabel('Yield of Apples')
```

<h5> Adding multiple lines </h5>

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Years')
plt.ylabel('Yield over the years')
```

<h5> Chart Title and Legend </h5>

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Years')
plt.ylabel('Yield over the years')
plt.title('Crops yield in Kanto')
plt.legend(['Apples', 'Oranges'])
```

<h5> Changing figure size </h5>

```python
plt.figure(figsize=(4,4))
plt.plot(years, oranges)
plt.title('Yield of Oranges')
```

<h5> Improving default Styles using Seaborn </h5>

```python
sns.set_style('darkgrid')
```

```python
plt.plot(years, apples, 's--b')
plt.plot(years, oranges, 'o--r')
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.title('Crop yields in Kanto')
plt.legend(['Apples', 'Oranges'])
```
